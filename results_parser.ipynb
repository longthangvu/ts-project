```python
import pandas as pd
from IPython.display import display

# 1) Load the aggregated results
df = pd.read_csv('results.csv')

    
pivot = df.pivot_table(
    index='model',
    columns='data',
    values='mse',
    aggfunc='mean'
)

# c) (optional) sort datasets alphabetically or by a custom order:
#    Here we sort alphabetically; you can supply your own list if you prefer
pivot = pivot.reindex(sorted(pivot.columns), axis=1)

# d) display with three decimals and a caption
display(
    pivot.style
            .format("{:.3f}")
            .set_caption(f"Average MSE")
)
```

```python
# 1) Load your aggregated results
df = pd.read_csv('results.csv')

# 2) Pivot on both data and pred_len
pivot = df.pivot_table(
    index='model',
    columns=['data','pred_len'],   # < multilevel cols
    values='mse',
    aggfunc='mean'
)

# 3) Sort columns by dataset name and then numeric pred_len
pivot = pivot.sort_index(axis=1, level=[0,1])

# 4) (Optional) flatten the multiindex into single labels:
# pivot.columns = [f"{d}_pl{p}" for d,p in pivot.columns]

# 5) Display with three decimals
display(
    pivot.style
         .format("{:.3f}")
         .set_caption("Average MSE by Dataset and Prediction Length")
)
```
